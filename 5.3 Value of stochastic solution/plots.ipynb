```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# Stacked bar plot (Fig. 8)
rcParams.update({'font.family':"serif", "font.serif": ["helvetica"]})
x = ["Stochastic solution", "Deterministic solution"]
y1 = np.array([2769610, 2685277.6])
y2 = np.array([4531624, 4524443.438])
y3 = np.array([4224865, 4238343.065])
y4 = np.array([0, 227079.69])
plt.bar(x, y1, label = "Module cost", width = 0.33, color = "#357fae")
plt.bar(x, y2, bottom = y1, label = "Production cost", width = 0.33, color = "#c7dbd5")
plt.bar(x, y3, bottom = y1 + y2, label = "Transportation cost", width = 0.33, color = "#4ea59f")
plt.bar(x, y4, bottom = y1 + y2 + y3, label = "Slack cost", width = 0.33, color = "#1e592b")
plt.text("Stochastic solution", y1[0] + y2[0] + y3[0] + 200000, f"Total: ${y1[0] + y2[0] + y3[0]:.0f}", ha = "center")
plt.text("Deterministic solution", y1[1] + y2[1] + y3[1] + y4[1] + 200000, f"Total: ${y1[1] + y2[1] + y3[1] + y4[1]:.0f}", ha = "center")
plt.legend()

fdict = {"family": "Helvetica", "weight": "bold", "fontsize": 12}
plt.ylim([0, 1.25e7])
plt.xlabel("Mobility option", fontdict = fdict)
plt.ylabel("Cost ($)", fontdict = fdict)
plt.title("")
```
